Objective:
The objective is to explore the data to identify the pattern that causes the person to have a good or bad or standard type credit score and build a machine learning model that should be able to predict or classify the credit score type.


```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the csv file
df = pd.read_csv("/content/train.csv")
```

```python
#Read the test csv file
df1=pd.read_csv("/content/test.csv")
df1.head()
```

```python
# Create a new column to identify the source of each row
df['Source'] = 'train'
df1['Source'] = 'test'

# Concatenate the two dataframes
merged_df = pd.concat([df, df1], axis=0, ignore_index=True)

merged_df.head()
```

```python
merged_df["Occupation"].value_counts()
```

```python
#Replacing abnormalities with the values closet to mean and mode of the dataset of occupation column
merged_df['Occupation'] = merged_df['Occupation'].replace(['_______', 'Other'], 'Developer')
```

```python
merged_df["Occupation"].value_counts()
```

```python
merged_df["SSN"].value_counts()
```

```python
merged_df[merged_df["Occupation"] == "Developer"]["SSN"].value_counts()
```

```python
#Replacing the SSN abnormalities with the mode of the SSN column
merged_df['SSN'] = merged_df['SSN'].replace(['#F%$D@*&8', 'Other'], '867-10-9609')
```

```python
merged_df["SSN"].value_counts()
```

```python
merged_df["Payment_Behaviour"].value_counts()
```

```python
#Replacing the Payment behaviour data abnormalities with its mode value
merged_df["Payment_Behaviour"]=merged_df["Payment_Behaviour"].replace(['!@9#%8','other'],'Low_spent_Small_value_payments')
```

```python
merged_df["Payment_Behaviour"].value_counts()
```

```python
#One hot encoding for Payment_Behaviour column in binary mode
merged_df = pd.get_dummies(merged_df, columns=['Payment_Behaviour'], prefix=['Payment_Behaviour'])
```

```python
merged_df.head()
```

```python
#Group by 'Customer_ID' and calculate the median age for each customer
customer_median_age = merged_df.groupby('Customer_ID')['Age'].apply(lambda x: pd.to_numeric(x, errors='coerce').median())

#Function to replace abnormal age values with customer-wise median
def replace_abnormal_age(row):
  if row['Age'] > 100 or row['Age'] < 18:
    return customer_median_age[row['Customer_ID']]
  else:
    return row['Age']

#Apply the function to the 'Age' column
merged_df['Age'] = pd.to_numeric(merged_df['Age'], errors='coerce') #Converting the 'Age' column to numeric
merged_df['Age'] = merged_df.apply(replace_abnormal_age, axis=1)
```

```python
# List of columns that should not have negative values
columns_no_negative = ['Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date', 'Num_Credit_Inquiries', 'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month']

# Loop through the columns and replace negative values with customer-wise median
for column in columns_no_negative:
  # Group by 'Customer_ID' and calculate the median for each customer
  customer_median = merged_df.groupby('Customer_ID')[column].apply(lambda x: pd.to_numeric(x, errors='coerce').median())

  # Function to replace negative values with customer-wise median
  def replace_negative_with_median(row):
    if pd.to_numeric(row[column], errors='coerce') < 0:
      return customer_median[row['Customer_ID']]
    else:
      return row[column]

  # Apply the function to the column
  merged_df[column] = merged_df.apply(replace_negative_with_median, axis=1)

# Check for other abnormalities in each variable
for column in merged_df.columns:
  # Check for unrealistic values or unusual patterns
  if merged_df[column].dtype == 'object':
    # For categorical variables, check for inconsistent values or typos
    print(f"Column: {column}, Unique Values: {merged_df[column].unique()}")
  else:
    # For numerical variables, check for outliers or unrealistic values
    print(f"Column: {column}, Descriptive Statistics: \n{merged_df[column].describe()}")
```

```python
merged_df.isnull().sum()
```

```python
# List of columns to handle missing values with customer-wise median
columns_with_missing = ['Age', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Delay_from_due_date', 'Num_Credit_Inquiries', 'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Total_EMI_per_month']

for column in columns_with_missing:
  # Group by 'Customer_ID' and calculate the median for each customer
  # Convert the column to numeric type before calculating the median
  customer_median = merged_df.groupby('Customer_ID')[column].transform(lambda x: pd.to_numeric(x, errors='coerce').median())

  # Fill missing values with customer-wise median
  merged_df[column].fillna(customer_median, inplace=True)

# Verify if missing values are handled
print(merged_df.isnull().sum())
```

```python
import re

def convert_credit_history_age(age_str):
  if pd.isnull(age_str):
    return np.nan
  match = re.match(r'(\d+)\s+years\s+and\s+(\d+)\s+months', age_str)
  if match:
    years = int(match.group(1))
    months = int(match.group(2))
    return years + (months / 12)
  else:
    return np.nan

merged_df['Credit_History_Age'] = merged_df['Credit_History_Age'].apply(convert_credit_history_age)

# Convert 'Payment_of_Min_Amount' to binary
# Replace 'NM' with 0 before converting to integers.
merged_df['Payment_of_Min_Amount'] = merged_df['Payment_of_Min_Amount'].replace(['No', 'Yes', 'NM'], [0, 1, 0])
merged_df['Payment_of_Min_Amount'] = merged_df['Payment_of_Min_Amount'].fillna(0).astype(int)

# Convert Month to numeric representation
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}

merged_df['Month'] = merged_df['Month'].map(month_mapping)
```

```python
merged_df.info()
```

```python
# Calculate the mode of 'Num_of_Delayed_Payment'
mode_delayed_payment = merged_df['Num_of_Delayed_Payment'].mode()[0]

# Fill null values with the mode
merged_df['Num_of_Delayed_Payment'].fillna(mode_delayed_payment, inplace=True)
```

```python
#Making a list of columns where replacing of abnormalities are required and datatype conversion to 'int' is required
num_col = ['Num_of_Delayed_Payment', 'Num_of_Loan']
for col in num_col:
  merged_df[col] = merged_df[col].str.replace('_','')
  # Convert to numeric, handling errors by coercing to NaN
  merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')
  # Fill NaN values with 0 and then convert to integers
  merged_df[col] = merged_df[col].fillna(0).astype(int)
```

```python
#Making a list of columns where replacing of abnormalities are required and datatype conversion to 'float' is required
float_col = ['Amount_invested_monthly', 'Monthly_Balance','Annual_Income','Outstanding_Debt']
for col in float_col:
  merged_df[col] = merged_df[col].str.replace('_','').astype(float)
```

```python
#Changed_Credit_Limit column abnormalities '_' are to be replaced by '0' and datatype conversion to 'float'
merged_df['Changed_Credit_Limit'] = merged_df['Changed_Credit_Limit'].str.replace('_', '0').astype(float)
```

```python
#'Num_Credit_Inquiries' column abnormalities '_' are to be replaced with '' and datatype conversion to float
merged_df['Num_Credit_Inquiries'] = merged_df['Num_Credit_Inquiries'].replace('_', '').astype(float)
```

```python
#one hot encoding for credit mix and removing '_' column for better model training
merged_df = pd.get_dummies(merged_df, prefix=['Credit_Mix'], columns=['Credit_Mix'])
merged_df = merged_df.drop('Credit_Mix__', axis=1)
```

```python
#Dropping Credit_History_Age column as it is nill
merged_df = merged_df.drop('Credit_History_Age', axis=1)
```

```python
merged_df['Credit_Score'].value_counts()
```

```python
#Label Encoding for Credit_Score column, poor = 0, standard = 1, good = 2
credit_score_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
merged_df['Credit_Score'] = merged_df['Credit_Score'].map(credit_score_mapping)
```

```python
merged_df['Credit_Score'].value_counts()
```

```python
merged_df.info()
```

```python
merged_df.isnull().sum()
```

```python
#Handling Amount_invested_monthly & Monthly_Balance columns null values with mode of the column
merged_df['Amount_invested_monthly'] = merged_df['Amount_invested_monthly'].fillna(merged_df['Amount_invested_monthly'].mode()[0])
merged_df['Monthly_Balance'] = merged_df['Monthly_Balance'].fillna(merged_df['Monthly_Balance'].mode()[0])
```

```python
merged_df.isnull().sum()
```

```python
# After EDA,we separate the dataframes
train_df = merged_df[merged_df['Source'] == 'train'].drop('Source', axis=1)
test_df = merged_df[merged_df['Source'] == 'test'].drop('Source', axis=1)
```
